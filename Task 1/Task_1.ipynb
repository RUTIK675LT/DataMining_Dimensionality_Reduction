PCA

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the dataset
file_path = '/content/pca_best_case.csv'  # Replace with the correct path in Colab
df = pd.read_csv(file_path)

# Standardizing the data is important for PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Apply PCA
pca = PCA(n_components=2)  # Reducing to 2 dimensions for visualization
X_pca = pca.fit_transform(X_scaled)

# Visualization
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.title('PCA Dimensionality Reduction')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import TruncatedSVD
import umap
import seaborn as sns

# Load the dataset
url = '/content/svd_example_dataset.csv'  # Replace with your CSV file URL
data = pd.read_csv(url)

# Perform SVD
n_components = 2  # You can vary this number
svd = TruncatedSVD(n_components=n_components)
svd_data = svd.fit_transform(data)

# Visualization of Singular Values
plt.figure(figsize=(8, 4))
plt.bar(range(1, n_components + 1), svd.singular_values_)
plt.xlabel('Singular Value Index')
plt.ylabel('Singular Value Magnitude')
plt.title('Singular Values')
plt.show()

# Visualization of the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(svd_data[:, 0], svd_data[:, 1], alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('First two Principal Components from SVD')
plt.grid(True)
plt.show()

# Hyperparameter Selection: Choosing the number of components
explained_variance = np.cumsum(svd.explained_variance_ratio_)
plt.figure(figsize=(8, 4))
plt.plot(explained_variance)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance vs. Number of Components')
plt.show()

# Advanced Visualization with UMAP
umap_model = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2)
umap_results = umap_model.fit_transform(data)

# Plotting the UMAP results
plt.figure(figsize=(10, 8))
sns.scatterplot(x=umap_results[:, 0], y=umap_results[:, 1], palette='viridis', legend='full')
plt.title('UMAP Visualization of the Dataset')
plt.xlabel('UMAP Component 1')
plt.ylabel('UMAP Component 2')
plt.grid(True)
plt.show()
```

LLE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import LocallyLinearEmbedding
from umap import UMAP

# Load the dataset
df = pd.read_csv('/content/simple_dataset_for_LLE.csv')

# Extract features
X = df.values

# Perform LLE
def perform_lle(n_neighbors, n_components):
    lle = LocallyLinearEmbedding(n_neighbors=n_neighbors, n_components=n_components, random_state=42)
    X_transformed = lle.fit_transform(X)
    return X_transformed

# Function to visualize the results using UMAP
def visualize_with_umap(X_transformed):
    umap = UMAP(random_state=42)
    embedding = umap.fit_transform(X_transformed)

    plt.scatter(embedding[:, 0], embedding[:, 1])
    plt.title('UMAP Visualization of LLE-transformed Data')
    plt.xlabel('UMAP1')
    plt.ylabel('UMAP2')
    plt.show()

# Set the parameters for LLE
n_neighbors = 10  # You can adjust this
n_components = 2  # You can adjust this

# Perform LLE with the set parameters and visualize
X_transformed = perform_lle(n_neighbors, n_components)
visualize_with_umap(X_transformed)
```

UMAP

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from umap import UMAP

# Load the dataset
df = pd.read_csv('/content/simple_dataset_for_UMAP.csv')

# Extract features
X = df.values

# Perform UMAP
def perform_umap(n_neighbors, n_components, min_dist):
    umap_model = UMAP(n_neighbors=n_neighbors, n_components=n_components, min_dist=min_dist, random_state=42)
    X_transformed = umap_model.fit_transform(X)
    return X_transformed

# Function to visualize the results
def visualize_umap(X_transformed, title):
    plt.scatter(X_transformed[:, 0], X_transformed[:, 1])
    plt.title(title)
    plt.xlabel('UMAP1')
    plt.ylabel('UMAP2')
    plt.show()

# Set the parameters for UMAP
n_neighbors = 15  # This parameter controls how UMAP balances local versus global structure in the data
n_components = 2  # The number of dimensions to project down to
min_dist = 0.1    # The effective minimum distance between embedded points

# Perform UMAP with the set parameters and visualize
X_transformed = perform_umap(n_neighbors, n_components, min_dist)
visualize_umap(X_transformed, 'UMAP Visualization with n_neighbors={}, n_components={}, min_dist={}'.format(n_neighbors, n_components, min_dist))
```

t-SNE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Load the dataset
df = pd.read_csv('/content/simple_dataset_for_tSNE.csv')

# Extract features
X = df.values

# Perform t-SNE
def perform_tsne(perplexity, n_components):
    tsne = TSNE(n_components=n_components, perplexity=perplexity, random_state=42)
    X_transformed = tsne.fit_transform(X)
    return X_transformed

# Function to visualize the results
def visualize_tsne(X_transformed, title):
    plt.scatter(X_transformed[:, 0], X_transformed[:, 1])
    plt.title(title)
    plt.xlabel('t-SNE1')
    plt.ylabel('t-SNE2')
    plt.show()

# Set the parameters for t-SNE
perplexity = 30  # Perplexity is a key parameter and can significantly impact the result. Typical values are between 5 and 50.
n_components = 2  # The number of dimensions to reduce to

# Perform t-SNE with the set parameters and visualize
X_transformed = perform_tsne(perplexity, n_components)
visualize_tsne(X_transformed, 't-SNE Visualization with perplexity={}'.format(perplexity))
```

Best and Worst cases of the following dimensionality reduction techniques -

    PCA (Principal Component Analysis)
        Best Case: A dataset with linear relationships where PCA can successfully capture most of the variance in the first few principal components. The visualization would show clear separation and variance along the principal component axes.
        Worst Case: A nonlinear dataset, like a Swiss roll. The PCA visualization would fail to unroll it, showing a compressed, overlapped representation without revealing the intrinsic structure.

    SVD (Singular Value Decomposition)
        Best Case: A dataset with strong linear correlations where SVD identifies and emphasizes these correlations. The visualization would show data points aligned along singular vectors, highlighting the underlying linear relationships.
        Worst Case: Data with significant noise or outliers. The SVD visualization would be dominated by these outliers, obscuring the true data structure.

    MDS (Multidimensional Scaling)
        Best Case: A dataset where distances/distortions can be well-represented in lower-dimensional space. The MDS visualization would accurately reflect the distances between points.
        Worst Case: Data requiring non-Euclidean distances for accurate representation. The MDS plot might show distorted structures and fail to maintain true distances.

    ISOMap (Isometric Mapping)
        Best Case: A dataset with a clear nonlinear manifold structure, like a Swiss roll, where ISOMap successfully unrolls it. The visualization would show a well-unfolded structure.
        Worst Case: Data with disconnected neighborhoods. ISOMap visualization would inaccurately connect these neighborhoods, distorting the true structure.

    LLE (Locally Linear Embedding)
        Best Case: Data with uniformly dense regions where LLE can accurately reconstruct each point from its neighbors. The visualization would show a well-preserved local structure.
        Worst Case: Data with varying densities or noise. LLE would struggle, leading to a visualization where some clusters might be overly compressed or expanded.

    UMAP (Uniform Manifold Approximation and Projection)
        Best Case: A dataset with complex local structures that UMAP can effectively unfold. The visualization would show clear, distinct local clusters.
        Worst Case: Data where global relationships are key. UMAP might produce a visualization that emphasizes local similarities at the expense of global structure.

    t-SNE (t-Distributed Stochastic Neighbor Embedding)
        Best Case: High-dimensional data with distinct clusters. t-SNE would create a visualization with well-separated clusters, making it easy to identify groupings.
        Worst Case: Data with important global structures or uniform distributions. t-SNE might either overemphasize cluster separations or create arbitrary clusters, misleading the interpretation.
